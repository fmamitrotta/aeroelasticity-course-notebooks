## Derivation of the Equations of Motion

Lagrange's equations:

$$\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial \mathscr{L}}{\partial \dot{\bm{q}}}\right) - \frac{\partial \mathscr{L}}{\partial \bm{q}} = \bm{Q},$$

where $\mathscr{L} = T - V$ is the Lagrangian, $T$ is the kinetic energy, $V$ is the potential energy, $\bm{q}$ are the generalized coordinates, $\bm{Q}$ are the non-conservative generalized forces.

The potential energy of the system is given by:

$$V = \frac{1}{2}k_h h^2 + \frac{1}{2}k_\theta \theta^2.$$

The total kinetic energy of the system is given by the sum of the translational and rotational kinetic energies:

$$T = \frac{1}{2}m\bm{v}_C \cdot \bm{v}_C + \frac{1}{2}I_C\dot{\theta}^2,$$

where $m$ is the mass of the airfoil, $\bm{v}_C$ is the velocity vector of the center of mass, and $I_C$ is the mass moment of inertia about the center of mass. The velocity vector of the center of mass is given by the vector sum of the plunge velocity, which is evaluated at point P, and of the tangential velocity due to pitch:

$$\bm{v}_C = \bm{v}_P + \bm{v}_t.$$

The plunge velocity is given by:

$$\bm{v}_P = -\dot{h}\hat{\bm{i}}_2,$$

while the tangential velocity due to pitch is given by the cross product of the angular velocity vector and the position vector from the rotation point P to the center of mass C:

$$\bm{v}_t = \theta\hat{\bm{b}}_3 \times b\left((1+a) - (1+e)\right)\hat{\bm{b}}_1 = b\dot{\theta}(a - e)\hat{\bm{b}}_2,$$

where $b$ is the semi-chord length, $a$ is the non-dimensional distance from point P to the aerodynamic center Q, and $e$ is the non-dimensional distance from point P to the center of mass C.

The unit vector $\hat{\bm{b}}_2$ has components in the inertial frame given by $\cos\theta\hat{\bm{i}}_2$ and $-\sin\theta\hat{\bm{i}}_1$. Consequently, the velocity vector of the center of mass can be expressed in the inertial frame as:

$$\bm{v}_C = -\dot{h}\hat{\bm{i}}_2 + b\dot{\theta}(a - e)(\cos\theta\hat{\bm{i}}_2 - \sin\theta\hat{\bm{i}}_1).$$

To simplify the expression, we assume $\theta$ to be small, allowing us to approximate $\cos\theta \approx 1$ and $\sin\theta \approx \theta$. Thus, the velocity vector of the center of mass simplifies to:

$$\bm{v}_C \approx \left(-\dot{h} + b\dot{\theta}(a - e)\right)\hat{\bm{i}}_2 - b\dot{\theta}(a - e)\theta\hat{\bm{i}}_1.$$

The second term can be neglected in the assumption of small $\theta$, and consequently we can write the product $\bm{v}_C \cdot \bm{v}_C$ as:

$$\bm{v}_C \cdot \bm{v}_C \approx \left(-\dot{h} + b\dot{\theta}x_\theta\right)^2 = \dot{h}^2 - 2bx_\theta\dot{h}\dot{\theta} + b^2x_\theta^2\dot{\theta}^2,$$

where we have defined $x_\theta = a - e$ for brevity.

Substituting this expression into the kinetic energy equation, we obtain:

$$T = \frac{1}{2}m\left(\dot{h}^2 - 2bx_\theta\dot{h}\dot{\theta} + b^2x_\theta^2\dot{\theta}^2\right) + \frac{1}{2}I_C\dot{\theta}^2 = \frac{1}{2}m\left(\dot{h}^2 - 2bx_\theta\dot{h}\dot{\theta}\right) + \frac{1}{2}I_P\dot{\theta}^2,$$

where $I_P = I_C + mb^2x_\theta^2$ is the mass moment of inertia about point P, obtained using the parallel axis theorem.

The Lagrangian $\mathscr{L} = T - V$ can now be expressed as:

$$\mathscr{L} = \frac{1}{2}m\left(\dot{h}^2 - 2bx_\theta\dot{h}\dot{\theta}\right) + \frac{1}{2}I_P\dot{\theta}^2 - \frac{1}{2}k_h h^2 - \frac{1}{2}k_\theta \theta^2.$$

We then take the partial derivatives of the Lagrangian with respect to the generalized coordinates and their time derivatives:

$$
\frac{\partial \mathscr{L}}{\partial h} = -k_h h, \quad \frac{\partial \mathscr{L}}{\partial \theta} = -k_\theta \theta,\\[1ex]
\frac{\partial \mathscr{L}}{\partial \dot{h}} = m\dot{h} - mbx_\theta \dot{\theta}, \quad \frac{\partial \mathscr{L}}{\partial \dot{\theta}} = -mbx_\theta \dot{h} + I_P \dot{\theta}.
$$

Successively, we compute the time derivatives $\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial \mathscr{L}}{\partial \dot{\bm{q}}}\right)$:

$$
\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial \mathscr{L}}{\partial \dot{h}}\right) = m\ddot{h} - mbx_\theta \ddot{\theta},\\[1ex]
\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\partial \mathscr{L}}{\partial \dot{\theta}}\right) = -mbx_\theta \ddot{h} + I_P \ddot{\theta}.
$$

The generalized forces associated with the degrees of freedom $h$ and $\theta$ can be derived from the work done by the aerodynamic forces and moments through virtual displacements and rotations:

$$\bm{Q} = \frac{\partial (\delta W)}{\partial (\delta \bm{q})}.$$

The work done by the aerodynamic forces is given by the sum of the work done by the lift force $L$ through the virtual displacement of point Q, $\delta\bm{p}_Q$, and the work done by the pitching moment about the same point, $M_{\frac{1}{4}}$, through the virtual rotation $\delta\theta$:

$$\delta W = L\hat{\bm{i}}_2 \cdot \delta\bm{p}_Q + M_{\frac{1}{4}}\hat{\bm{i}}_3 \cdot \delta\theta\hat{\bm{i}}_3.$$

Using thin-airfoil steady-flow theory, the lift force and pitching moment about the quarter-chord can be expressed as:

$$
L = \frac{1}{2}\rho U^2 c c_l = 2\pi \rho b U^2 \theta\\[1ex]
M_{\frac{1}{4}} = 0
$$

where $c=2b$ is the chord length and $c_l = 2\pi\theta$ is the lift coefficient for a thin airfoil at small angle of attack.

The virtual displacement vector of point Q is obtained similarly to the velocity vector of point C, as the sum of the plunge virtual displacement at point P and of the tangential virtual displacement at point Q due to pitch:

$$\delta\bm{p}_Q = -\delta h\hat{\bm{i}}_2 + \delta\theta\hat{\bm{b}}_3 \times b\left((1+a) - \frac{1}{2}\right)\hat{\bm{b}}_1 = -\delta h\hat{\bm{i}}_2 + b\delta\theta(\frac{1}{2} + a)\hat{\bm{b}}_2.$$

Substituting the expressions for $L$, $M_{\frac{1}{4}}$, and $\delta\bm{p}_Q$ into the work equation, and using the small-angle approximations for $\hat{\bm{b}}_2$, we obtain:

$$\delta W = 2\pi \rho b U^2 \theta \left(-\delta h + b\delta\theta\left(\frac{1}{2} + a\right)\right),$$

and taking the partial derivatives with respect to $\delta h$ and $\delta \theta$, we find the generalized forces:

$$
Q_h = \frac{\partial (\delta W)}{\partial (\delta h)} = -2\pi \rho b U^2 \theta = -L,\\[1ex]
Q_\theta = \frac{\partial (\delta W)}{\partial (\delta \theta)} = 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)\theta = b\left(\frac{1}{2} + a\right)L.
$$

Substituting the expressions for the Lagrangian derivatives and the generalized forces into Lagrange's equations, we obtain the equations of motion for the typical section with pitch and plunge degrees of freedom:

$$
m\ddot{h} - mbx_\theta \ddot{\theta} + k_h h = -2\pi \rho b U^2 \theta,\\[1ex]
-mbx_\theta \ddot{h} + I_P \ddot{\theta} + k_\theta \theta = 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)\theta.
$$

We note that this is a system of two coupled second-order linear ordinary differential equations with constant coefficients and that we can rewrite it in homogeneous form by bringing all terms to the left-hand side:

$$
m\ddot{h} - mbx_\theta \ddot{\theta} + k_h h + 2\pi \rho b U^2 \theta = 0,\\[1ex]
-mbx_\theta \ddot{h} + I_P \ddot{\theta} + k_\theta \theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)\theta = 0.
$$

## Flutter analysis with the p method

The **p method** is an eigenvalue-based approach for analyzing the flutter stability of aeroelastic systems. It is one of the most fundamental and widely used techniques in flutter analysis because it provides complete information about the system's dynamic behavior: not only the flutter boundary (where instability begins), but also the modal frequencies and damping characteristics at all flight speeds.

The p method is based on three key principles:

1. **Exponential solution form**: For linear, time-invariant systems with constant coefficients, the general solution to the homogeneous differential equations can be expressed as a sum of exponential functions.

2. **Eigenvalue analysis**: By substituting the exponential solution into the equations of motion, we transform the differential equation problem into an algebraic eigenvalue problem. The eigenvalues determine the stability and oscillatory characteristics of the system.

3. **Nondimensionalization**: By introducing dimensionless variables, we can express the solution in terms of a small number of fundamental parameters that characterize the aeroelastic system, making the results more general and easier to interpret.

The p method is particularly valuable because it provides modal damping information at subcritical speeds (below flutter), which is essential for assessing safety margins, and it identifies which modes are becoming unstable and how they couple.

### Assumed solution form

For a linear system of ordinary differential equations with constant coefficients, the general solution can be expressed as a superposition of exponential functions. Therefore, we seek solutions of the form:

$$h(t) = \overline{h} e^{\nu t}, \quad \theta(t) = \overline{\theta} e^{\nu t},$$

where $\nu$ is a complex constant (to be determined) and $\overline{h}$ and $\overline{\theta}$ are complex amplitudes (also to be determined). $\nu$ is has units of [1/time] and it tells us "how fast things happen in absolute time". By writing $\nu$ in terms of its real and imaginary parts:

$$\nu = \zeta + i\omega,$$

we can understand the physical meaning of its components:

1. **$\omega$ - Circular frequency**: The imaginary part $\omega$ (rad/s) is the **circular frequency of oscillation**. It determines how rapidly the system oscillates in time. The oscillation period is $T = 2\pi/\omega$ and the frequency in Hz is $f = \omega/(2\pi)$.

2. **$\zeta$ - Damping rate**: The real part $\zeta$ (1/s) is the **exponential growth or decay rate** of the amplitude. The sign of $\zeta$ determines the stability:
   - **$\zeta < 0$**: Amplitude decays exponentially → **Stable** (damped motion)
   - **$\zeta = 0$**: Amplitude remains constant → **Neutrally stable** (simple harmonic motion)
   - **$\zeta > 0$**: Amplitude grows exponentially → **Unstable** (flutter)

Computing the time derivatives:

$$\dot{h} = \nu \overline{h} e^{\nu t}, \quad \ddot{h} = \nu^2 \overline{h} e^{\nu t},$$
$$\dot{\theta} = \nu \overline{\theta} e^{\nu t}, \quad \ddot{\theta} = \nu^2 \overline{\theta} e^{\nu t}.$$

Substituting these expressions into the equations of motion:

$$
m\nu^2 \overline{h} e^{\nu t} - mbx_\theta \nu^2 \overline{\theta} e^{\nu t} + k_h \overline{h} e^{\nu t} + 2\pi \rho b U^2 \overline{\theta} e^{\nu t} = 0,\\[1ex]
-mbx_\theta \nu^2 \overline{h} e^{\nu t} + I_P \nu^2 \overline{\theta} e^{\nu t} + k_\theta \overline{\theta} e^{\nu t} - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)\overline{\theta} e^{\nu t} = 0.
$$

Since $e^{\nu t} \neq 0$ for all $t$, we can divide it out from both equations, yielding:
$$
(m\nu^2 + k_h) \overline{h} + (-mbx_\theta \nu^2 + 2\pi \rho b U^2) \overline{\theta} = 0,\\[1ex]
(-mbx_\theta \nu^2) \overline{h} + \left(I_P \nu^2 + k_\theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)\right) \overline{\theta} = 0.
$$

In matrix form, this becomes:

$$
\begin{bmatrix}
m\nu^2 + k_h & -mbx_\theta \nu^2 + 2\pi \rho b U^2 \\[1ex]
-mbx_\theta \nu^2 & I_P \nu^2 + k_\theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)
\end{bmatrix}
\begin{Bmatrix}\overline{h} \\[1ex] \overline{\theta}\end{Bmatrix} =
\begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix},
$$

which is a system of homogeneous linear algebraic equations. Thus by using the assumed exponential solution form, we have transformed the original system of ordinary differential equations into a system of algebraic equations.

### Eigenvalue problem formulation

Now we want to rewrite the system in order to separate the terms involving $\nu^2$ from the constant terms. We then introduce the mass and stiffness matrices of the aeroelastic system:

$$\bm{M} = \begin{bmatrix} m & -mbx_\theta \\ -mbx_\theta & I_P \end{bmatrix}, \quad \bm{K} = \begin{bmatrix} k_h & 2\pi \rho b U^2 \\ 0 & k_\theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right) \end{bmatrix}.$$

Note that $\bm{K}$ contains both structural stiffness terms and aerodynamic terms that depend on the flight speed $U$. The system can now be written as:

$$(\nu^2 \bm{M} + \bm{K}) \begin{Bmatrix}\overline{h} \\ \overline{\theta}\end{Bmatrix} = \begin{Bmatrix}0 \\ 0\end{Bmatrix}.$$

This is a **quadratic eigenvalue problem** where $\nu$ is the eigenvalue and $\overline{\bm{q}} = \{\overline{h}, \overline{\theta}\}^T$ is the eigenvector. For nontrivial solutions to exist (both $\overline{h}$ and $\overline{\theta}$ nonzero), we require:

$$\det(\nu^2 \bm{M} + \bm{K}) = 0.$$

This determinant equation is a polynomial in $\nu^2$. Since our system has two degrees of freedom, this is a quadratic equation in $\nu^2$, which yields two values of $\nu^2$. Taking both positive and negative square roots of each, we obtain **four values of $\nu$**:

$$\nu_1, \nu_2, \nu_3, \nu_4.$$

These four values are the **eigenvalues** of our aeroelastic system, and in general they appear as two complex conjugate pairs since the physical motion $h(t)$ and $\theta(t)$ are real-valued functions:

$$\nu_{1,2} = \zeta_1 \pm i\omega_1, \quad \nu_{3,4} = \zeta_2 \pm i\omega_2.$$

Each complex conjugate pair corresponds to a distinct mode of motion, where $\zeta$ is called the **modal damping rate** and $\omega$ is called the **modal frequency**. The general solution of the system is a linear combination of the four exponential functions corresponding to each eigenvalue:

$$\bm{q}(t) = \sum_{k=1}^{4} c_k \overline{\bm{q}}_k e^{\nu_k t},$$

or explicitly for each degree of freedom:

$$h(t) = \sum_{k=1}^{4} c_k \overline{h}_k e^{\nu_k t}, \quad \theta(t) = \sum_{k=1}^{4} c_k \overline{\theta}_k e^{\nu_k t},$$

where $c_k$ are constants determined by initial conditions, and $\overline{\bm{q}}_k = \{\overline{h}_k, \overline{\theta}_k\}^T$ is the eigenvector corresponding to eigenvalue $\nu_k$. From these last expressions, we can see that if one of the eigenvalues has a positive real part ($\zeta > 0$), the motion will grow without bound, indicating **flutter instability**.

To determine the flutter speed with the p method, we need to analyze how the eigenvalues $\nu$ change as a function of flight speed $U$. The flutter boundary occurs when the real part of one of the eigenvalues crosses zero ($\zeta = 0$), indicating a transition from stable to unstable behavior.

To find the eigenvalues as a function of $U$, we can either solve the determinant equation directly for each value of $U$ or reformulate the quadratic eigenvalue problem into a standard eigenvalue problem by rewriting the system in state-space form.

By defining the state vector $\bm{x}$ as:
$$\bm{x} = \begin{Bmatrix} \bm{q} \\ \dot{\bm{q}} \end{Bmatrix},$$

we can rewrite our second order system $\bm{M}\ddot{\bm{q}} + \bm{K}\bm{q} = \mathbf{0}$ as a first-order system:

$$\dot{\bm{x}} = \bm{A}\bm{x},$$

where:
$$\bm{A} = \begin{bmatrix} \mathbf{0} & \mathbf{I} \\ -\mathbf{M}^{-1}\mathbf{K} & \mathbf{0} \end{bmatrix}.$$

If we assume $\bm{x} = \overline{\bm{x}}e^{\nu t}$:
$$\nu\overline{\bm{x}}e^{\nu t} = \bm{A}\overline{\bm{x}}e^{\nu t},$$

and cancel $e^{\nu t}$, we find the standard eigenvalue problem:
$$\bm{A}\overline{\bm{x}} = \nu\overline{\bm{x}}.$$

The exponential time constants $\nu$ are the eigenvalues of the state matrix $\bm{A}$, and $\overline{\bm{x}}$ are the eigenvectors, which contain both the modal amplitudes and velocities.

Thus, to solve the flutter problem, we need to assemble the state matrix $\bm{A}$ for various values of flight speed $U$, compute its eigenvalues $\nu$, and track how their real parts change with $U$. The flutter speed $U_F$ is identified as the value of $U$ where the real part of one eigenvalue crosses zero.

### Nondimensionalization and dimensionless parameters

To generalize our analysis and reduce the number of parameters, we introduce dimensionless variables. We begin by introducing the **uncoupled natural frequencies** at zero airspeed:

$$\omega_h = \sqrt{\frac{k_h}{m}}, \quad \omega_\theta = \sqrt{\frac{k_\theta}{I_P}},$$

which represent the frequencies at which the system would oscillate if the plunge and pitch degrees of freedom were completely uncoupled and there were no aerodynamic forces.

Next, we introduce the **dimensionless eigenvalue** $p$, whence the name of the method, defined as:

$$p = \frac{b\nu}{U} = \frac{b}{U}\left(\zeta + i\omega\right).$$

The quantity $Ut/b$ is the **number of semi-chords traveled by the flow** in time $t$. Therefore the dimensionless eigenvalue $p$ characterizes **how the motion evolves per semi-chord of flow travel**, rather than per unit of time.

Next, we introduce a dimensionless velocity. We define the **reduced velocity $V$** as:

$$V = \frac{U}{b\omega_\theta}.$$

This quantity represents **how many semi-chords the flow travels during one period of pitch oscillation** and it is the **primary independent variable** in flutter analysis. As we increase $V$, we track how the system's stability changes.

To complete the nondimensionalization, we introduce three more dimensionless parameters that fully characterize the aeroelastic system:

1. **Mass ratio**, $\mu$:
   $$\mu = \frac{m}{\pi\rho b^2}$$
   
   This represents the ratio of the airfoil mass per unit span to the mass of a cylinder of air with radius $b$ and unit span.

2. **Radius of gyration squared**, $r^2$:
   $$r^2 = \frac{I_P}{mb^2}$$
   
   This represents the ratio of the moment of inertia of the airfoil about the rotation point P to the moment of inertia of a point mass with the same mass of the airfoil located at a distance $b$ from the rotation axis.

3. **Frequency ratio**, $\sigma$:
   $$\sigma = \frac{\omega_h}{\omega_\theta}$$
   
   This compares the uncoupled plunge and pitch natural frequencies.

We can now rewrite our system in terms of these dimensionless parameters. Starting from:

$$
\begin{bmatrix}
m\nu^2 + k_h & -mbx_\theta \nu^2 + 2\pi \rho b U^2 \\[1ex]
-mbx_\theta \nu^2 & I_P \nu^2 + k_\theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)
\end{bmatrix}
\begin{Bmatrix}\overline{h} \\[1ex] \overline{\theta}\end{Bmatrix} =
\begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix},
$$

we substitute $\nu = pU/b$:

$$
\begin{bmatrix}
m\frac{p^2U^2}{b^2} + k_h & -m b x_\theta \frac{p^2U^2}{b^2} + 2\pi \rho b U^2 \\[1ex]
-m b x_\theta \frac{p^2U^2}{b^2} & I_P \frac{p^2U^2}{b^2} + k_\theta - 2\pi \rho b^2 U^2 \left(\frac{1}{2} + a\right)
\end{bmatrix}
\begin{Bmatrix}\overline{h} \\[1ex] \overline{\theta}\end{Bmatrix} =
\begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix}.
$$

Successively, in order to have both equations in the same units, we multiply the first equation by $b$ and we multiply and divide each term associated to $\overline{h}$ by $b$, so that we can work with $\overline{h}/b$ instead of $\overline{h}$. Next we also divide we divide both equations by $mU^2$:

$$
\begin{bmatrix}
p^2 + \frac{k_h b^2}{mU^2} & -x_\theta p^2 + \frac{2\pi \rho b^2}{m} \\[1ex]
-x_\theta p^2 & \frac{I_P p^2}{mb^2} + \frac{k_\theta}{mU^2} - \frac{2\pi \rho b^2}{m}\left(\frac{1}{2} + a\right)
\end{bmatrix}
\begin{Bmatrix}\frac{\overline{h}}{b} \\[1ex] \overline{\theta}\end{Bmatrix} =
\begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix}.
$$

Next, we express each term using our dimensionless parameters. Using the definitions:

$$
U = V b \omega_\theta, \quad k_h = \omega_h^2 m, \quad k_\theta = \omega_\theta^2 I_P, \quad \mu = \frac{m}{\pi\rho b^2}, \quad r^2 = \frac{I_P}{mb^2}, \quad \sigma = \frac{\omega_h}{\omega_\theta},
$$

we can rewrite each term:

- $\frac{k_h b^2}{mU^2} = \frac{\omega_h^2 m b^2}{m V^2 b^2 \omega_\theta^2} = \frac{\sigma^2}{V^2}$

- $\frac{2\pi \rho b^2}{m} = \frac{2}{\mu}$

- $\frac{I_P}{mb^2} = r^2$

- $\frac{k_\theta}{mU^2} = \frac{\omega_\theta^2 I_P}{m V^2 b^2 \omega_\theta^2} = \frac{r^2}{V^2}$

- $\frac{2\pi \rho b^2}{m} = \frac{2}{\mu}$

Substituting these expressions into our matrix equation yields:

$$
\begin{bmatrix}
p^2 + \frac{\sigma^2}{V^2} & -x_\theta p^2 + \frac{2}{\mu} \\[1ex]
-x_\theta p^2 & r^2 p^2 + \frac{r^2}{V^2} - \frac{2}{\mu}\left(a + \frac{1}{2}\right)
\end{bmatrix}
\begin{Bmatrix}\frac{\overline{h}}{b} \\[1ex] \overline{\theta}\end{Bmatrix} =
\begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix}.
$$

By introducing the nondimensional mass and stiffness matrices:

$$\bm{M}^* = \begin{bmatrix} 1 & -x_\theta \\ -x_\theta & r^2 \end{bmatrix}, \quad \bm{K}^* = \begin{bmatrix} \frac{\sigma^2}{V^2} & \frac{2}{\mu} \\ 0 & \frac{r^2}{V^2} - \frac{2}{\mu}\left(a + \frac{1}{2}\right) \end{bmatrix},$$

we can rewrite the system as a quadratic eigenvalue problem in dimensionless form:

$$\left(p^2 \bm{M}^* + \bm{K}^*\right) \begin{Bmatrix}\frac{\overline{h}}{b} \\[1ex] \overline{\theta}\end{Bmatrix} = \begin{Bmatrix}0 \\[1ex] 0\end{Bmatrix}.$$

This dimensionless form reveals that the flutter behavior depends only on:
- Five configuration parameters: $\mu$, $r$, $\sigma$, $a$, $x_\theta$
- One flight condition parameter: $V$ (the reduced velocity)

Analogously to the dimensional case, to find the eigenvalues $p$ we can either solve the determinant equation directly or reformulate the problem in state-space form. To reformulate the problem in state-space form we introduce the nondimensional state vector:

$$\bm{x}^* = \begin{Bmatrix} \bm{q}^* \\ \dot{\bm{q}}^* \end{Bmatrix},$$

where $\bm{q}^* = \{\overline{h}/b, \overline{\theta}\}^T$. With this, we can rewrite the second order system $\bm{M}^*\ddot{\bm{q}}^* + \bm{K}^*\bm{q}^* = \mathbf{0}$ as a first-order system:

$$\dot{\bm{x}}^* = \bm{A}^*\bm{x}^*,$$

where $\bm{A}^*$ is the nondimensional state matrix:

$$\bm{A}^* = \begin{bmatrix} \mathbf{0} & \mathbf{I} \\ -\mathbf{M}^{*-1}\mathbf{K}^* & \mathbf{0} \end{bmatrix}.$$

By assembling $\bm{A}^*$ for various values of $V$, we can compute its eigenvalues $p$ and track how their real parts change with $V$. The flutter reduced velocity $V_F$ is identified as the value of $V$ where the real part of one eigenvalue crosses zero.

Since $\bm{A}^*$ is a $4 \times 4$ matrix, we obtain four eigenvalues, analogously to the dimensional case:

$$p_1, \, p_2, \, p_3, \, p_4,$$

which appear as two complex conjugate pairs:

$$p_{1,2} = \frac{b}{U}\left(\zeta_1 \pm i\omega_1\right), \quad p_{3,4} = \frac{b}{U}\left(\zeta_2 \pm i\omega_2\right).$$

### Computational Procedure for Flutter Analysis

To determine the flutter characteristics using the p method, we calculate the eigenvalues $p$ for a range of reduced velocities and plot their real and imaginary parts to visualize the trend with $V$. However, the eigenvalues $p$ contain the flight speed $U$ in the denominator, which means that both their real and imaginary parts tend to increase as $U$ decreases and viceversa. Consequently, we can't rely directly on the real and imaginary parts of $p$ to assess the damping and 

For this purpose, it is convenient to present the real and imaginary parts of the eigenvalues $p$ multiplied by $V$:

$$p \cdot V = \frac{b}{U}(\zeta + i\omega) \cdot \frac{U}{b\omega_\theta} = \frac{\zeta}{\omega_\theta} + i\frac{\omega}{\omega_\theta}.$$

In this way, the real and imaginary parts of the calculated eigenvalues $p$ do not depend on the flight speed $U$, 

The **flutter boundary** is defined as the flight condition at which one of the modes transitions from stable to unstable. Mathematically, this occurs when:

$$\Gamma = 0 \quad \text{for one of the modes}.$$

At this critical condition, we define:

1. **Flutter reduced velocity**, $V_F$:
   The value of $V$ at which $\Gamma = 0$ for the first time as $V$ is increased from zero.

2. **Flutter speed**, $U_F$:
   $$U_F = V_F \cdot b\omega_\theta.$$

3. **Flutter frequency**, $\omega_F$:
   $$\omega_F = \Omega_F V_F \omega_\theta,$$
   where $\Omega_F$ is the imaginary part of the eigenvalue at the flutter condition.

4. **Flutter mode shape**:
   The eigenvector $\{\overline{h}/b, \overline{\theta}\}_F$ corresponding to the flutter eigenvalue determines the relative amplitudes and phase between plunge and pitch motions at flutter.

**Physical significance**: Below the flutter speed ($V < V_F$), all modes have $\Gamma < 0$ and are stable. At the flutter speed ($V = V_F$), one mode has $\Gamma = 0$ and exhibits simple harmonic motion with constant amplitude. Above the flutter speed ($V > V_F$), that mode has $\Gamma > 0$ and becomes exponentially divergent, leading to structural failure.

To generate the modal frequency and damping curves, we follow these computational steps:

1. **Specify configuration parameters**:
   - Choose values for: $a$ (aerodynamic center location), $e$ (center of mass location), or equivalently $x_\theta = e - a$
   - Choose: $\mu$ (mass ratio), $r$ (radius of gyration), $\sigma$ (frequency ratio)

2. **Create velocity array**:
   - Generate an array of $V$ values: $V = [V_{\min}, V_{\min} + \Delta V, \ldots, V_{\max}]$
   - Start near zero (e.g., $V_{\min} = 0.1$) to avoid singularities
   - Extend beyond expected flutter point (e.g., $V_{\max} = 3.0$ or higher)
   - Use sufficient resolution (e.g., $\Delta V \approx 0.01$) to capture details

3. **For each value of $V$**:
   - Assemble the nondimensional state matrix $\bm{A}^*$
   - Compute its eigenvalues $p_1, p_2, p_3, p_4$
   - Extract modal frequencies $\Omega_k$ (imaginary parts) and modal damping $\Gamma_k$ (real parts)

4. **Store and organize results**:
   - Group eigenvalues by conjugate pairs
   - Track which modes correspond to which physical characteristics (e.g., plunge-dominated vs pitch-dominated)

5. **Generate plots**:
   - **Modal Frequency**: Plot $\Omega$ versus $V$ for each mode
   - **Modal Damping**: Plot $\Gamma$ versus $V$ for each mode

6. **Identify flutter point**:
   - Find the value of $V$ where $\Gamma$ first crosses zero
   - Determine the corresponding $V_F$, $\omega_F$, and mode shape

### Implementation: Calculation of Flutter Curves

We now implement the computational procedure described above to reproduce the flutter analysis results from Hodges and Pierce (2011). We will use the same configuration parameters as in the example from Section 5.2 of the book:

- $a = -1/5$ (aerodynamic center location)
- $e = -1/10$ (center of mass location)
- $\mu = 20$ (mass ratio)
- $r^2 = 6/25$ (radius of gyration squared)
- $\sigma = 2/5$ (frequency ratio)

From these we can derive $x_\theta = e - a = -1/10 - (-1/5) = 1/10$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

# Configuration parameters (from Hodges & Pierce, Section 5.2)
a = -1/5           # Aerodynamic center location
e = -1/10          # Center of mass location
x_theta = a - e    # Static unbalance parameter
mu = 20            # Mass ratio
r_squared = 6/25   # Radius of gyration squared
r = np.sqrt(r_squared)
sigma = 2/5        # Frequency ratio

print(f"Configuration parameters:")
print(f"  a = {a}")
print(f"  e = {e}")
print(f"  x_theta = {x_theta}")
print(f"  μ = {mu}")
print(f"  r² = {r_squared} → r = {r:.6f}")
print(f"  σ = {sigma}")
```

Next, we define a function to assemble the dimensionless state matrix $\bm{A}^*$ for a given reduced velocity $V$:

```python
def assemble_state_matrix(V, mu, r, sigma, a, x_theta):
    """
    Assemble the dimensionless state matrix A* for the typical section flutter problem.
    
    Parameters:
    -----------
    V : float
        Reduced velocity V = U/(b*omega_theta)
    mu : float
        Mass ratio μ = m/(π*ρ*b²)
    r : float
        Radius of gyration r = sqrt(I_P/(m*b²))
    sigma : float
        Frequency ratio σ = omega_h/omega_theta
    a : float
        Dimensionless aerodynamic center location
    x_theta : float
        Static unbalance parameter x_theta = e - a
        
    Returns:
    --------
    A_star : ndarray (4x4)
        Dimensionless state matrix
    """
    
    # Dimensionless mass matrix M*
    M_star = np.array([
        [1.0,     -x_theta],
        [-x_theta, r**2   ]
    ])
    
    # Dimensionless stiffness matrix K*
    K_star = np.array([
        [sigma**2/V**2,  2.0/mu                        ],
        [0.0,            r**2/V**2 - 2.0*(a + 0.5)/mu  ]
    ])
    
    # Inverse of mass matrix
    M_star_inv = linalg.inv(M_star)
    
    # State matrix A* = [0  I; -M*^(-1)*K*  0]
    zeros = np.zeros((2, 2))
    eye = np.eye(2)
    
    A_star = np.block([
        [zeros,              eye        ],
        [-M_star_inv @ K_star, zeros      ]
    ])
    
    # Return the assembled state matrix
    return A_star
```

We now sweep through a range of reduced velocities and compute the eigenvalues at each point:

```python
# Create array of reduced velocities
V_min = 0.1
V_max = 3.0
n_points = 300
V_array = np.linspace(V_min, V_max, n_points)

# Arrays to store results
# We have 2 complex conjugate eigenvalue pairs at each V
Omega_array = np.zeros((n_points, 4))  # Modal frequencies
Gamma_array = np.zeros((n_points, 4))  # Modal damping

# Compute eigenvalues for each V
for i, V in enumerate(V_array):
    # Assemble state matrix
    A_star = assemble_state_matrix(V, mu, r, sigma, a, x_theta)
    
    # Compute eigenvalues
    eigenvalues = linalg.eigvals(A_star)*V

    # Sort eigenvalues by imaginary part (frequency)
    idx = np.argsort(np.abs(eigenvalues.imag))
    eigenvalues = eigenvalues[idx]

    # Check that eigenvalues come in conjugate pairs, otherwise print warning
    if not np.isclose(eigenvalues[0], np.conj(eigenvalues[1])) or not np.isclose(eigenvalues[2], np.conj(eigenvalues[3])):
        print(f"Warning: Eigenvalues at V={V:.3f} do not form conjugate pairs as expected.")
    
    # Extract Gamma (real part) and Omega (imaginary part)
    Gamma_array[i, :] = eigenvalues.real
    Omega_array[i, :] = eigenvalues.imag

print(f"\nEigenvalue computation complete.")
print(f"Computed eigenvalues for {n_points} values of V from {V_min} to {V_max}")
```

Let's plot the modal frequencies and damping ratios as functions of the reduced velocity to visualize the flutter characteristics of the system.

```python
# Enable interactive plots
%matplotlib widget

# Plot modal frequencies and damping ratios
_, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Modal Frequencies
for i in range(np.size(Omega_array, 1)):
    axes[0].plot(V_array, Omega_array[:, i], 'o')
axes[0].set_ylabel('Modal Frequencies $\Omega$')
axes[0].grid(True)

# Modal Damping
for i in range(np.size(Gamma_array, 1)):
    axes[1].plot(V_array, Gamma_array[:, i], 'o')
axes[1].set_ylabel('Modal Damping $\Gamma$')
axes[1].set_xlabel('Reduced Velocity $V$')
axes[1].grid(True)

# Set tight layout and show plots
plt.tight_layout()
plt.show()
```
