# Theodorsen's Theory for Unsteady Aerodynamics

![Theodorsen's airfoil-aileron combination.](resources/theodorsen-airfoil.png)

Hybrid expression of the lift defined in the time-frequency domain:

$$
L\left(t, k\right) = L_\mathrm{nc}\left(t\right) + L_\mathrm{c}\left(t, k\right).
$$

Non-circulatory component of the lift:

$$
L_\mathrm{nc}\left(t\right)=\pi\rho b^2\left(V\dot{\alpha}\left(t\right) + \ddot{h}_0\left(t\right) - ab\ddot{\alpha}\left(t\right)\right).
$$

Circulatory component of the lift:

$$
L_\mathrm{c}\left(t, k\right) = 2\pi\rho b V C\left(k\right) \left(V\alpha\left(t\right) + \dot{h}_0\left(t\right) + b\left(\frac{1}{2} - a\right)\dot{\alpha}\left(t\right)\right).
$$

Theodorsen function:

$$
C\left(k\right) = \dfrac{H_1^{\left(2\right)}\left(k\right)}{H_1^{\left(2\right)}\left(k\right) + i H_0^{\left(2\right)}\left(k\right)}.
$$

Calculate Theodorsen function for a reduced frequency range between 0 and 10.

```python
import numpy as np
from scipy.special import hankel2

k_step = 1e-3
k_final = 10
k_vector = np.arange(k_step, k_final + k_step, k_step)  # k = 0 is not included because it produces a nan
theodorsen = hankel2(1, k_vector)/(hankel2(1, k_vector) + 1j*hankel2(0, k_vector))
```

Plot magnitude and phase of Theodorsen function.

```python
from matplotlib import pyplot as plt

magnitude = np.abs(theodorsen)
phase = np.angle(theodorsen)  # [rad]
fig, axes = plt.subplots(2, 1)
axes[0].plot(k_vector, magnitude)
axes[0].set_ylabel('Magnitude')
axes[0].grid()
axes[1].plot(k_vector, np.rad2deg(phase))  # convert phase to degrees
axes[1].set_ylabel('Phase, deg')
axes[1].set_xlabel('$k$')
axes[1].grid()
plt.show()
```
